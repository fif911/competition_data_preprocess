```python
import pandas as pd

df = pd.read_csv('V-Dem-CY-Full+Others-v13.csv')

df
```

```python
remains = ['country_name','country_text_id','year','e_gdp','e_gdppc']

df = df[remains]

df
```

```python
df = df[df["year"] >= 1970]
df = df[df['year'] <= 2018]
df
```

```python
# Separate West Germany data
df_germany = df[df['country_name'] == 'Germany']
df_west_germany = df_germany[df_germany['year'] < 1991]  # West Germany existed till 1990
df_west_germany['country_name'] = 'West Germany'
df_west_germany['country_text_id'] = 'DEU'  

# Update original Germany data
df_germany = df_germany[df_germany['year'] >= 1991]

df_germany
```

```python
# Do the same for Soviet Union and Russia
df_russia = df[df['country_name'] == 'Russia']
df_soviet_union = df_russia[df_russia['year'] < 1991]  # Soviet Union existed till 1991
df_soviet_union['country_name'] = 'Soviet Union'
df_soviet_union['country_text_id'] = 'SUN'  # replace with actual code if different

# Update original Russia data
df_russia = df_russia[df_russia['year'] >= 1991]
df_russia
```

```python
# Now you can concat these dataframes to your original data, after removing the old Germany and Russia data
df = df[(df['country_name'] != 'Germany') & (df['country_name'] != 'Russia')]

df = pd.concat([df, df_west_germany, df_germany, df_soviet_union, df_russia])
```

```python
# Separate vietnam data
df_serbia = df[df['country_name'] == 'Serbia']
df_yugoslavia_serbia = df_serbia[(df_serbia['year'] >= 1960) & (df_serbia['year'] <= 1999)]
df_yugoslavia_serbia.loc[:, 'country_name'] = 'Yugoslavia'
df_yugoslavia_serbia.loc[:, 'country_text_id'] = 'SFR'  # replace with actual code if different

# Update original Serbia data
df_serbia = df_serbia[df_serbia['year'] > 1999]
df_yugoslavia_serbia
```

```python
# Do the same for Montenegro
df_montenegro = df[df['country_name'] == 'Montenegro']
df_yugoslavia_montenegro = df_montenegro[(df_montenegro['year'] >= 1999) & (df_montenegro['year'] <= 2002)]
df_yugoslavia_montenegro.loc[:, 'country_name'] = 'Yugoslavia'
df_yugoslavia_montenegro.loc[:, 'country_text_id'] = 'SFR'  # replace with actual code if different

# Update original Montenegro data
df_montenegro = df_montenegro[df_montenegro['year'] > 2002]
```

```python
# Now you can concat these dataframes to your original data, after removing the old Serbia and Montenegro data
df = df[(df['country_name'] != 'Serbia') & (df['country_name'] != 'Montenegro')]

df = pd.concat([df, df_yugoslavia_serbia, df_serbia, df_yugoslavia_montenegro, df_montenegro])
```

```python
df.to_csv('gdp_pc.csv',index=False)
```
